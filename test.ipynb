```python
import pandas as pd
import os
from backtest import Backtest
import matplotlib.pyplot as plt
```

```python
# 数据路径为作业要求的默认路径
data_path = os.path.join('..','data','stk_daily.feather')
# 请输入需要回测的范围
start_date = pd.Timestamp(year=2021,month=10,day=1)
end_date = pd.Timestamp(year=2022,month=11,day=5)
# 请输入n日截面反转策略的n的值，起始金额
n = 5
cash = 10000

# 初始化大约需要20~30分钟
backtest = Backtest(data_path,start_date,end_date,n,cash)
backtest.run_trading()
asset_curve_list,excess_return_rate,annualized_return_rate,annualized_volatility,sharpe_ratio,maximun_drawdown = backtest.output()
```

```python
# 绘制曲线
plt.plot(asset_curve_list)
# 添加标题和标签
plt.title('Equity Curve')
plt.xlabel('date')
plt.ylabel('total asset')
# 显示曲线
plt.show()
print(f'超额回报率为：{excess_return_rate}')
print(f'年化收益率为：{annualized_return_rate}')
print(f'年化波动为：{annualized_volatility}')
print(f'夏普比率为：{sharpe_ratio}')
print(f'最大回撤率为：{maximun_drawdown}')
```
